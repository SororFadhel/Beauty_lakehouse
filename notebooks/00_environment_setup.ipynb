# Environment Initialization for the Beauty Lakehouse

This setup creates the shared Unity Catalog environment used by all team members. It is executed once by the project owner or workspace admin before running any ingestion or analytics notebooks.

## Purpose of Each Object

### Catalog
A catalog provides the top-level namespace for all project data. Creating `beauty_catalog` ensures that curated, warehouse, and analytics tables are stored in a governed, shared location accessible to the entire team.

### Schemas
Schemas organize tables by layer within the Lakehouse:
- `curated` — cleaned Delta tables produced by ingestion  
- `warehouse` — fact and dimension tables  
- `analytics` — KPI and reporting outputs  

This structure keeps the project modular and consistent.

### Volume
A volume provides a governed storage location for files such as raw data and curated Delta outputs. Creating `workspace.beauty.data` gives the team a shared path:



```python
%sql
CREATE CATALOG IF NOT EXISTS workspace;
CREATE SCHEMA IF NOT EXISTS workspace.beauty;
CREATE VOLUME IF NOT EXISTS workspace.beauty.data;

CREATE CATALOG IF NOT EXISTS beauty_catalog;
CREATE SCHEMA IF NOT EXISTS beauty_catalog.curated;
CREATE SCHEMA IF NOT EXISTS beauty_catalog.warehouse;
CREATE SCHEMA IF NOT EXISTS beauty_catalog.analytics;
```

## Environment Setup Complete

The Unity Catalog environment is now ready.

You may proceed to **01_dataLake_ingestion**, where raw data will be loaded, validated, and saved into the shared Volume and catalog tables.
